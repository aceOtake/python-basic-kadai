```python
# 課題：繰り返し処理を使いこなそう
array = ["水","金","地","火","木","土","天","海","冥"]

# forループで水～冥の9行を表示
print('--- forループで水～冥の9行を表示 ---')
for item in array:
  print(item)

# whileループで水～冥の9行を表示
print('--- whileループで水～冥の9行を表示 ---')

# インデックスの最大値を取得
max_index = len(array) - 1
# ループカウンタ
counta = 0
while max_index >= counta:
  print(array[counta])
  counta += 1
```
